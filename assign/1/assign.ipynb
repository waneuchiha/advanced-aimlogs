## Άσκηση 1: Μαθηματικές εκφράσεις

Ένα αντικείμενο αποτελείται από έναν κύλινδρο και δύο ημισφαίρια ίδιας διαμέτρου προσαρμοσμένα στις βάσεις του.
Γράψτε μια συνάρτηση με το όνομα `myfunction1`, η οποία που δέχεται ως παράμετρο την διάμετρο $d$ και υπολογίζει τον λόγο όγκου προς επιφάνεια $\frac{V}Α$ του αντικειμένου.
Χρησιμοποιείστε την σταθερά $π$ της μονάδας κώδικα numpy.
Για τo ύψος του κυλίνδρου $h$ χρησιμοποιείστε την σχέση:

$$h=a+\frac{b+1}{c+1}$$

όπου $a,b,c$ είναι τα 3 τελευταία ψηφία του αριθμού μητρώου σας ($c$=τελευταίο).

Ελέγξτε τα αποτελέσματα της συνάρτησης για $d=1.0$, $d=2.5$ και $d=5.0$.

```python
import numpy as np

def myfunction1(d: np.float64) -> np.float64:
    # Τρία τελευταία ψηφία του ΑΜ: 167
    a = 1
    b = 6
    c = 7

    # Υπολογισμός ύψους κυλίνδρου
    h = a + (b + 1) / (c + 1)

    # Ακτίνα
    r = d / 2.0

    # Όγκος: κύλινδρος + σφαίρα (δύο ημισφαίρια)
    V = np.pi * r**2 * h + (4/3) * np.pi * r**3

    # Επιφάνεια: κύλινδρος + σφαίρα
    A = 2 * np.pi * r * h + 4 * np.pi * r**2

    # Λόγος όγκου προς επιφάνεια
    VA = V / A

    return np.float64(VA)

```

```python
# Πρόχειρο άσκησης 1
# - Σε αυτό το κελί μπορείτε να τυπώσετε μεταβλητές για επαλήθευση. Δεν λαμβάνεται υπόψη στην βαθμολόγηση.

print(myfunction1(1.0), myfunction1(2.5), myfunction1(5.0))
```

## Άσκηση 2: Πίνακες NumPy
Γράψτε μια συνάρτηση `myfunction2`, η οποία δέχεται ως παράμετρο ένα διάνυσμα $P_1$ δύο στοιχείων (μονοδιάστατος πίνακας numpy.ndarray με συντεταγμένες $x_1$ και $y_2$) και επιστρέφει την γωνία $\theta$ που σχηματίζει το διάνυσμα $P_1$ με το διάνυσμα:

$P_2=\begin{bmatrix} x_2\\ y_2\end{bmatrix}=\begin{bmatrix} 2b-c\\ a+3\end{bmatrix}$

Τα $a,b,c$ είναι τα 3 τελευταία ψηφία του αριθμού μητρώου σας ($c$=τελευταίο).

Για τον υπολογισμό χρησιμοποιείστε τον νόμο των συνημιτόνων:

$$
\cos \theta =\frac{P_1\cdot P_2}{|P_1|\cdot|P_2|}
$$

όπου $P_1\cdot P_2=x_1x_2+y_1y_2$ το εσωτερικό γινόμενο και $|P_1|,|P_2|$ τα μέτρα των διανυσμάτων.

Υπόδειξη: Η γωνία $\theta$ να υπολογιστεί σε ακτίνια στο πεδίο $[0,\pi]$.

Ελέγξτε τα αποτελέσματα της συνάρτησης για $P_1=\begin{bmatrix} 0\\ 8\end{bmatrix}$ και $P_1=\begin{bmatrix} 5\\ 2\end{bmatrix}$.

```python

def myfunction2(P1: np.ndarray) -> np.float64:
    # Τρία τελευταία ψηφία του ΑΜ: 167
    a = 1
    b = 6
    c = 7

    # Ορισμός διανύσματος P2
    P2 = np.array([2*b - c, a + 3], dtype=np.float64)

    # Εσωτερικό γινόμενο
    dot_product = np.dot(P1, P2)

    # Μέτρα διανυσμάτων
    norm_P1 = np.linalg.norm(P1)
    norm_P2 = np.linalg.norm(P2)

    # Υπολογισμός cos(θ)
    cos_theta = dot_product / (norm_P1 * norm_P2)

    # Διόρθωση πιθανών αριθμητικών σφαλμάτων (ώστε να μην ξεφύγει από [-1,1])
    cos_theta = np.clip(cos_theta, -1.0, 1.0)

    # Υπολογισμός γωνίας σε ακτίνια
    theta = np.arccos(cos_theta)

    return np.float64(theta)
```

```python
# Πρόχειρο άσκησης 2
# - Σε αυτό το κελί μπορείτε να τυπώσετε μεταβλητές για επαλήθευση. Δεν λαμβάνεται υπόψη στην βαθμολόγηση.

print(myfunction2(np.array([0., 8.])), myfunction2(np.array([5., 2.])))
```
